```python
def encode(data: str, encoding: dict):
    out = ""
    for d in data:
        for k in encoding.keys():
            if d == k:
                out += encoding[k]
    return out
```

```python
def encode(node: dict):  
    codes = {}

    for c in node.keys():
        for direction in ["left", "right"]:
            if direction in c:
                if node["left_child"] == None:
                    codes[node["character"]] += str(0) + encode(node)
                elif node["right_child"] == None:
                    codes[node["character"]] += str(1) + encode(node)
                
                continue
    return codes

```
