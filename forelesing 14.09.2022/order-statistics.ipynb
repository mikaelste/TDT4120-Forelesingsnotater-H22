# 1. Sorteringsgrensen
&rarr; Hva kan vi garantere for en sorteringsalgoritmen?

Eksempel *Worst-case*:
- du har en permutasjon
-> av $n!$ mulige -> Trenger da maks $lgn!$ ja-nei-spørsmål

$T_w(n) = \Omega(lgn!) = \Omega(n*lgn) = "informasjonsbildet av en tilfeldig permutasjon" >= lgn$
-> $lgn$ er antall halveringer fra n! til 1
*En absolutt nedre grense for verste tilfelle*

Eksempel *Best-case*
- $T_b(n) = O(\infty)$
- $T_b(n) = \Theta(?)$
- $T_b(n) = \Omega(n)$

#### Sterlings approximasjon
$n! >= (n/e)^n$ for $T_w(n)$

# Randomisert select
- Hvem er på n-te plass?
-> Kan være bedre enn $n*lgn$

Bruker halvparten av quicksort (samme oppførsel som "quicksort som binærsøk")

- Instans
    - En ikke tom sekvens A[1:n] og et heltall i $\in$ [1:n]
    -> Det i-ende minste elementet i A
- Dekomponering
    - Fordel i to deler, venstre verdier mindre enn høyre
    -> Går tilbake til løsning i instans
- Delinstans
    - Venstre eller høyre del, A[p:r], og juster rang i'
    -> det i'--ende minste elementet i A[p:r]
- Grunntilfelle
    - Intervall av lengde 1, altså der p==r, og i' == 1
    -> A[p] (grunnløsningen)

Optimalisering -> Stopper umiddelbart etter funn

CASES:
- $T_w(n) = \Theta(n^2)$
- $T_A(n) = \Theta(n)$
- $T_B(n) = \Theta(n)$

# 4. Tellesortering

**Ikke-rekursiv** dekomponering (design)

- Instans

    - A[1:n] med verdiområde [0:k]

    -> A, sortert
    
- Dekomponering

    - Se på hver verdi for seg (hvor hører den hjemme)

    -> ... sett A[j] i posisjonen C[A[j]]

Delinstanser

    - Individuelle verdier A[j]

    -> Riktig posisjoner C[A[j]]

Effektivt fordi den først oppretter en liste med antallet av veriene som er i lista
-> bruker disse som indexer for å sette inn i ny liste B
-> Man krymper da verdien i antall-lista for at ny index for like verdier blir satt inn én før

KJØRETID :$\Theta(n+k)$

# 5. Radik (grunntall) - sortering

- Instans
    - En sekvens med d-sifrede tall
    -> A skal være sortert

- Dekomponering
    - Se på hvert siffer for seg

- Delinstanser
    - En sekvens for hvert siffer $i$
        - Bare en tolkning av a
    
    -> Samme sekvenser, stabilt sortert på stedet etter tur
    Hver sortering er en ny sortering av A

KJØRETID: $T(n) = \Theta(d*(n+k))$, der $(n+k)$ kommer fra stabil tellesortering

# 6. Bøttesortering

- Instans
    Sekvens A[1:n] med verdiområde [0, 1)
    -> A sortert
- Dekomponering
    - Fordel i n like intervaller langs verdiområdet
    -> Sett B[0], ..., B[n-1] sammen i stigende rekkefølge
- Delinstanser
    - En sekvens B[i] per intervall, i=1, ..., n-1
    -> B[0], ..., B[n-1], sorterte

Kjøretid:
- $T_w(n) = \Theta(n^2)$ -> Alle i samme bøtte
- $T_A(n) = \Theta(n)$
- $T_B(n) = \Theta(n)$
